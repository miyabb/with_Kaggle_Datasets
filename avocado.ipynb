# アボカドが有機なのか無機なのか予測する

```python
import pandas as pd

df = pd.read_csv('avocado.csv', index_col ='Date', parse_dates=['Date'])
```

```python
df.head(3)
```

```python
obj_col = 'type'
df[obj_col].value_counts()
```

```python
#ラベルを整数に変換

#行要素, 列要素 locはdataframeのlocation（変える部分を指定している）
#obj_colがconventionalなら0, organaicなら1
df.loc[df[obj_col]=='conventional',obj_col] = 0
df.loc[df[obj_col]=='organic',obj_col] = 1

#目的変数の型変換
df.loc[:,obj_col]=df.loc[:, obj_col].astype(int)
```

```python
exp_col = ['AveragePrice', 
           'Total Volume', 
           '4046', 
           '4225', 
           '4770', 
           'Total Bags', 
           'Small Bags', 
           'Large Bags', 
           'XLarge Bags']
```

```python
len(exp_col)
```

```python
df[exp_col].values
```

```python
df[obj_col].values
```

```python
train_df = df.loc[:'2017',:]
test_df = df.loc['2018':,:]
```

```python
train_df
```

```python
X_train = train_df.loc[:'2017', exp_col].values   #訓練　説明変数
y_train = train_df.loc[:'2017', obj_col].values   #訓練　目的変数（ラベル）

X_test = test_df.loc['2018':, exp_col].values     #テスト　説明変数
y_test = test_df.loc['2018':, obj_col].values     #テスト　目的変数（ラベル）
```

```python
test_df.loc[:,obj_col].value_counts()
```

## SVMで分類

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
#標準化させるために訓練データだけで平均と分散を計算
scaler.fit(X_train)
#訓練データ，テストデータを標準化する．
X_train = scaler.transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVC

clf = SVC(class_weight='balanced',random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print('平均予測精度 : {}\n'.format(clf.score(X_test, y_test)))
print(classification_report(y_test, y_pred,target_names=['conventional','organic']))
```

## Kerasのニューラルネットを使って予測

```python
from keras import models
from keras import layers

network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(len(exp_col),))) 
network.add(layers.Dense(10, activation='relu'))
network.add(layers.Dense(10, activation='relu'))
network.add(layers.Dense(10, activation='relu'))
network.add(layers.Dense(2, activation='softmax'))
```

```python
## ネットワークのコンパイル
## rmsprop = AdaGrad + 勾配の二乗の指数移動平均

network.compile(optimizer='Adam', 
                loss = 'binary_crossentropy',   #カテゴリカル変数（独立しているデータ（１・２・３など）離散的）
                metrics=['accuracy'])
```

```python
y_train.shape
```

```python
## ラベルエンコード => one-hot エンコード

from keras.utils import to_categorical

train_labels = to_categorical(y_train)
test_labels = to_categorical(y_test)
```

```python
train_labels.shape
```

```python
test_labels
```

```python
network.fit(X_train, train_labels, epochs=100, batch_size=1024)
```

```python
X_train
```

```python
test_loss, test_acc = network.evaluate(X_test, test_labels)
```
